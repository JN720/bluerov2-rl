```python
RESULT_FILE = 'training_results_2025-02-21_22:44:56.txt'
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
episode_rewards = open(RESULT_FILE).read().split('\n')
rewards = []
for episode_reward in episode_rewards:
    rewards.append(np.array([float(reward) for reward in episode_reward.split(',')[:-1]], dtype = np.float32))
```

```python
data = [np.mean(reward) for reward in rewards]
plt.plot(data)
```

```python
data = [np.std(reward) for reward in rewards]
plt.plot(data)
```

```python
data = [np.mean(np.abs(np.diff(reward))) for reward in rewards]
plt.plot(data)
```

```python
import os
import glob
files = sorted(glob.glob(os.path.join('data', "*")), key=os.path.getmtime, reverse=True)

newest_file_1, newest_file_2 = files[:2] if len(files) >= 2 else (files[0], None)
if 'center' in newest_file_1:
    DISTANCE_FROM_CENTER_FILE = newest_file_1
    DISTANCE_TO_GOAL_FILE = newest_file_2
else:
    DISTANCE_FROM_CENTER_FILE = newest_file_2
    DISTANCE_TO_GOAL_FILE = newest_file_1
# DISTANCE_TO_GOAL_FILE = 'data/distance_to_target_2025-02-24_22:20:17.txt'
# DISTANCE_FROM_CENTER_FILE = 'data/distance_from_center_2025-02-24_22:20:17.txt'
```

```python
goal_distances = open(DISTANCE_TO_GOAL_FILE).read().split('\n')
center_distances = open(DISTANCE_FROM_CENTER_FILE).read().split('\n')
goal_distances = [float(distance) for distance in goal_distances if distance != '']
center_distances = [float(distance.split(',')[1][:-1]) for distance in center_distances if distance != '']
```

```python
plt.plot(goal_distances)
plt.plot(center_distances)
```
